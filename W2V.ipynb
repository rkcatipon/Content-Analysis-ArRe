```
import time
import pickle
t0=time.time()
doc_sents=[]
dcount = 0
for (index, d) in newsDF.iterrows():
    sents = []
    for s in d['normalized_tokens']:
      t=[x.replace('medium','media') for x in s]
      sents.append(t)
    doc_sents.append(sents)
    dcount+=1
    if (dcount%10000==0):
      print('Processed', dcount ,'documents') 
newsDF['normalized_tokens'] = doc_sents
t1=time.time()
print('Time taken', t1-t0)
newsDF.to_pickle('/content/drive/My Drive/SOCI 40133 - Final Project/Data/full_news_df.pkl')
```
